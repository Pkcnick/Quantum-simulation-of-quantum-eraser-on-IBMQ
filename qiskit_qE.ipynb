```python
from qiskit import *
from numpy import pi
```

```python
qr = QuantumRegister(3, 'q')
```

```python
cr = ClassicalRegister(3, 'c')
```

```python
circuit = QuantumCircuit(qr , cr)
```

```python
circuit.draw()
```

```python
circuit.h(qr[0])
```

```python
circuit.draw()
```

```python
from qiskit.visualization import circuit_drawer
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.cx(qr[0] , qr[1])
```

```python
circuit.draw(output= 'mpl')
```

```python
circuit.x(qr[1])
```

```python
circuit.draw(output= 'mpl')
```

```python
circuit.s(qr[1])
circuit.h(qr[1])
circuit.s(qr[1])
circuit.z(qr[1])
circuit.y(qr[1])
circuit.cp(pi / 2, qr[2], qr[1])
circuit.cp(pi / 2, qr[2], qr[1])
circuit.x(qr[1])
circuit.s(qr[1])
circuit.h(qr[1])
circuit.s(qr[1])
```

```python
circuit.draw(output= 'mpl')
```

```python
circuit.cx(qr[0], qr[1])
circuit.h(qr[0])
```

```python
circuit.draw(output= 'mpl')
```

```python
circuit.z(qr[2])
circuit.y(qr[2])
```

```python
circuit.draw(output= 'mpl')
```

```python
circuit.p(pi / 2, qr[2])
```

```python
circuit.draw(output= 'mpl')
```

```python
circuit.s(qr[2])
circuit.h(qr[2])
circuit.s(qr[2])
```

```python
circuit.draw(output= 'mpl')
```

```python
import qiskit.quantum_info as qi
```

```python
circuit.measure(qr, cr)
```

```python
circuit.draw(output = 'mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(circuit, backend = simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(result.get_counts(circuit))
```

```python
IBMQ.load_account()
```

```python
provider = IBMQ.get_provider('ibm-q')
```

```python
qcomp = provider.get_backend('ibmq_qasm_simulator')
```

```python
job = execute(circuit, backend = qcomp, shots = 1000)
```

```python
from qiskit.tools.monitor import job_monitor
```

```python
job_monitor(job)
```

```python
result = job.result()
```
